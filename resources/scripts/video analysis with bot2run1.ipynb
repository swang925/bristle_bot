```python
import cv2 as cv
import os

bot_num = 2
run_num = 1

video = cv.VideoCapture("bot{}run{}.mov".format(bot_num, run_num))

folder_name = 'bot{}run{}'.format(bot_num, run_num)

try:
    #Create data folder
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
except OSError:
    print('Error: creating directory of frames of the video')
    
currentframe = 0

while True:
    ret,frame = video.read()
    if ret:
        # if not done reading, keep reading
        name = './{}/bot{}run{}frame{}.jpg'.format(folder_name, bot_num, run_num, currentframe)
        print('Creating...'+name)
        
        #writing the extracted images
        cv.imwrite(name,frame)
        
        currentframe += 1
    else:
        break
```
